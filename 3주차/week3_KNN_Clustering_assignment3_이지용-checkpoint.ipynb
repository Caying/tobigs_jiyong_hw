```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 데이터 불러오기

```python
data = pd.read_csv('Mall_Customers.csv', encoding='utf-8')

plt.style.use('seaborn')
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

# Preprocessing

### one hot encoding   
참고 : https://minjejeon.github.io/learningstock/2017/06/05/easy-one-hot-encoding.html

```python
data = data.drop(["CustomerID"], axis=1)
```

```python
gender_one_hot_encoded = pd.get_dummies(data.Gender) # 0 : Male, 1 : Female
data["Gender"] = gender_one_hot_encoded
data.head()
```

```python
gender = data['Gender'].value_counts() # value_counts는 클래스 별 갯수를 반환
plt.bar(gender.index,gender, color =['pink','skyblue']) # x는 인덱스, y는 빈도수, color로 색 지정
plt.xticks(gender.index, ['여자','남자']) # x축을 여자,남자 범주로 변환
plt.title('남녀 빈도 Bar Chart')
plt.xlabel('성별')
plt.ylabel('빈도')
plt.show()
```

# Hierarchical Clustering

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

df_gender = pd.DataFrame(data["Gender"])
df_gender.colums=['Gender']

# Calculate the linkage: mergings
mergings = linkage(data,method='complete')

# Plot the dendrogram, using varieties as labels
plt.figure(figsize=(40,20))
dendrogram(mergings,
           labels = df_gender.as_matrix(columns=['Gender']),
           leaf_rotation=90,
           leaf_font_size=10,
)

plt.show()
```

## Evaluate model with Cross tabulation

```python
from scipy.cluster.hierarchy import fcluster

predict = pd.DataFrame(fcluster(mergings,104,criterion='distance'))
predict.columns=['predict']
ct = pd.crosstab(predict['predict'],df_gender['Gender'])
print(ct)
```

# K-means Clustering

```python
feature = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans

# create model and prediction
model = KMeans(n_clusters=3,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)
```

```python
r.head()
```

```python
centers = pd.DataFrame(model.cluster_centers_,columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])
center_x = centers['Age']
center_y = centers['Annual Income (k$)']
center_z = centers['Spending Score (1-100)']

from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Age'],r['Annual Income (k$)'],r['Spending Score (1-100)'],c=r['predict'],alpha=1)
ax.scatter(center_x,center_y,center_z,s=50,marker='D',c='r')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

## Evaluate model with Cross tabulation

```python
ct = pd.crosstab(data['Gender'],r['predict'])
print (ct)
print (model.inertia_)
```

### Scaling

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
model = KMeans(n_clusters=3)
pipeline = make_pipeline(scaler,model)
pipeline.fit(feature)
predict = pd.DataFrame(pipeline.predict(feature))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)

ct = pd.crosstab(data['Gender'],r['predict'])
print (ct)
print (model.inertia_)
```

## 최적의 군집 수 K 결정 (Elbow Point)

```python
X = feature
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot  as plt
import seaborn as sns

# create model and prediction
model = DBSCAN(eps=0.3,min_samples=6)
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Age'],r['Annual Income (k$)'],r['Spending Score (1-100)'],c=r['predict'],alpha=1)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

## Evaluate model with Cross tabulation

```python
ct = pd.crosstab(data['Gender'],r['predict'])
print (ct)
```
