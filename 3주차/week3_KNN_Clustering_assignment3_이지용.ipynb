```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 데이터 불러오기

```python
data = pd.read_csv('Mall_Customers.csv', encoding='utf-8')

plt.style.use('seaborn')
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

# Preprocessing

### one hot encoding   
참고 : https://minjejeon.github.io/learningstock/2017/06/05/easy-one-hot-encoding.html

```python
data = data.drop(["CustomerID"], axis=1)
```

```python
gender_one_hot_encoded = pd.get_dummies(data.Gender) # 0 : Male, 1 : Female
data["Gender"] = gender_one_hot_encoded
data.head()
```

```python
gender_count = data['Gender'].value_counts() # value_counts는 클래스 별 갯수를 반환
plt.bar(gender_count.index,gender_count, color =['pink','skyblue']) # x는 인덱스, y는 빈도수, color로 색 지정
plt.xticks(gender_count.index, ['여자','남자']) # x축을 여자,남자 범주로 변환
plt.title('남녀 빈도 Bar Chart')
plt.xlabel('성별')
plt.ylabel('빈도')
plt.show()
```

### 상관관계

Age와 Spending Score 사이의 약한 음의 상관관계가 있다.

```python
g = sns.heatmap(data = data.corr(), annot=True, annot_kws={'fontsize':15}, fmt = '.2f', linewidths=.2, cmap='Blues')
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.show()
```

### 히스토그램

### 산점도(lmplot)

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

sc_data = scaler.fit(data)
sc_data = sc_data.transform(data)
sc_data_df = pd.DataFrame(sc_data)
sc_data_df.columns = data.columns
sc_data_df["Gender"] = gender_one_hot_encoded
sc_data_df.head()
```

# Hierarchical Clustering

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

df_gender = pd.DataFrame(sc_data_df["Gender"])
df_gender.colums=['Gender']

# Calculate the linkage: mergings
mergings = linkage(sc_data_df,method='complete')

# Plot the dendrogram, using varieties as labels
plt.figure(figsize=(40,20))
dendrogram(mergings,
           labels = df_gender.as_matrix(columns=['Gender']),
           leaf_rotation=90,
           leaf_font_size=10,
)

plt.show()
```

## Evaluate model with Cross tabulation

```python
from scipy.cluster.hierarchy import fcluster

predict = pd.DataFrame(fcluster(mergings,20,criterion='distance'))
predict.columns=['predict']
ct = pd.crosstab(predict['predict'],df_gender['Gender'])
print(ct)
```

# K-means Clustering

```python
feature = sc_data_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans

# create model and prediction
model = KMeans(n_clusters=6,algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)
```

```python
r.head()
```

```python
centers = pd.DataFrame(model.cluster_centers_,columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])
center_x = centers['Age']
center_y = centers['Annual Income (k$)']
center_z = centers['Spending Score (1-100)']

from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Age'],r['Annual Income (k$)'],r['Spending Score (1-100)'],c=r['predict'],alpha=1)
ax.scatter(center_x,center_y,center_z,s=50,marker='D',c='r')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

## Evaluate model

실루엣 스코어는 군집이 6개일때 가장 크게 나온다.

```python
from sklearn.metrics.cluster import silhouette_score

silhouette_score(feature, predict)
```

```python
ct = pd.crosstab(sc_data_df['Gender'],r['predict'])
print (ct)
print (model.inertia_)
```

## 최적의 군집 수 K 결정 (Elbow Point)

```python
X = feature
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

# create model and predition
model = DBSCAN(eps=0.575,min_samples=6)
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict']
```

```python
# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Age'],r['Annual Income (k$)'],r['Spending Score (1-100)'],c=r['predict'],alpha=1)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

X = feature

neigh = NearestNeighbors(n_neighbors=6)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

## Evaluate model

실루엣 스코어는 eps =575 일때 가장 크다

```python
silhouette_score(feature, predict)
```

```python
ct = pd.crosstab(sc_data_df['Gender'],r['predict'])
print (ct)
```

# SpectralClustering

In multivariate statistics and the clustering of data, spectral clustering techniques make use of the spectrum (eigenvalues) of the similarity matrix of the data to perform dimensionality reduction before clustering in fewer dimensions. The similarity matrix is provided as an input and consists of a quantitative assessment of the relative similarity of each pair of points in the dataset.  

In application to image segmentation, spectral clustering is known as segmentation-based object categorization.

https://en.wikipedia.org/wiki/Spectral_clustering

```python
from sklearn.cluster import SpectralClustering
```

```python
label = sc_data_df['Gender']
model = SpectralClustering(n_clusters=6, assign_labels="discretize", random_state=0)
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict']

r = pd.concat([feature, predict], axis = 1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Age'],r['Annual Income (k$)'],r['Spending Score (1-100)'],c=r['predict'],alpha=1)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

## Evaluate model

```python
silhouette_score(feature, predict)
```
