```python
import numpy as np
import pandas as pd
```

### 데이터 불러오기

```python
df = pd.read_csv('sampled_data.csv')
```

```python
df.head()
```

### Train Test 데이터 나누기

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() # Scaling
scaler.fit_transform(X_train)
```

```python
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
```

```python
X_train_scaled
```

Class를 y_train에서 다시 불러오기

```python
y_train_reset = y_train.reset_index()
y_train_reset = y_train_reset.drop(columns=["index"])
y_train_reset
```

```python
target = np.array(y_train_reset["Class"])
target
```

```python
X_train_scaled['Class'] = target # class 0: malignant, class 1: benign
X_train_scaled.head()
```

```python
#test도 스케일링
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test), columns = X_test.columns)
y_test_reset = y_test.reset_index()
y_test_reset["Class"]
target = np.array(y_test_reset["Class"])
X_test_scaled['Class'] = target # class 0: malignant, class 1: benign
X_test_scaled.head()
```

로지스틱 회귀 실행

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) # class 예측
y_pred
```

```python
classifier.predict_proba(X_test) # The returned estimates for all classes are ordered by the label of classes.
```

### mean accuracy 출력

```python
classifier.score(X_test, y_test) # return mean accuracy
```

Scaled Version으로 다시 classifying

```python
y_pred2 = classifier2.predict(X_test_scaled) # class 예측
y_pred2
```

```python
classifier2.score(X_test_scaled, y_test)
```

### ROC 커브 그리기

```python
fpr, tpr, thresholds = roc_curve(y_test, classifier.predict_proba(X_test)[:, 1])
# 순서대로 fpr, tpr, thresholds를 리턴함!
# fpr = 1-specificity
# tpr = sensitivity
# thresholds = 해당 fpr, tpr을 리턴할 때의 thesholds(cut-off) 값!
fpr, tpr, thresholds
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = fpr
y = tpr

plt.plot(x,y) # 간단하게 ROC 그려볼 수 있음.
plt.xlabel('1-Specificity')
plt.ylabel('Sensitivity')
```

```python
roc_auc_score(y_test, classifier.decision_function(X_test)) # auc_score 계산
```

```python
# sklearn 패키지에서 아예 plot 그리는 메소드를 제공 (0.22.1로 업데이트해야 함)

plot_roc_curve(classifier, X_test, y_test, response_method = 'decision_function')
```

```python
res = [[fpr_i, tpr_i, thres_i] for fpr_i, tpr_i, thres_i in zip(fpr, tpr, thresholds)]
```

```python
res # 이런 식으로 각 대응되는 값끼리 묶을 수 있습니다.
```

### 새로운 cut off 찾기

```python
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
```

```python
optimal_idx
```

```python
optimal_threshold
```

```python
proba = classifier.predict_proba(X_test)
proba
```

```python
newcut_predicted=[]
for prob in proba :
    if prob[1] > optimal_threshold:
        newcut_predicted.append(int(1))
    else :
        newcut_predicted.append(int(0))
newcut_predicted = np.array(newcut_predicted)
newcut_predicted
```

```python
confusion_matrix(y_test, newcut_predicted)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, newcut_predicted).ravel()
print(tn, fp, fn, tp)
```

```python
newcut_recall = tp/(fn+tp)
newcut_recall
```

```python
newcut_specificity = tn/(fp+tn)
newcut_specificity
```

```python
newcut_precision = tp/(fp+tp)
newcut_precision
```

```python
newcut_F1 = 2*newcut_precision*newcut_recall/(newcut_precision+newcut_recall)
newcut_F1
```
