# 인코딩

### 1) 서울/부산 구분

서울이면 0, 부산이면 1 이다.

```python
sido_bi = []
for idx, member in df.iterrows():
    if member["addr_si"] == "서울" :
        sido_bi.append(0)
    else :
        sido_bi.append(1)
    
df["sido_bi"] = sido_bi
df["sido_bi"]
```

### 2) 유찰 여부

유찰된 적이 없으면 0, 있으면 1 이다.

```python
miscarriage_bi = []
for idx, member in df.iterrows():
    if member["Auction_miscarriage_count"] == "0" :
        miscarriage_bi.append(0)
    else :
        miscarriage_bi.append(1)
    
df["miscarriage_bi"] = miscarriage_bi
df["miscarriage_bi"]
```

### 3) 건물면적과 건물경매면적의 일치여부

건물면적과 건물경매면적이 불일치하면 0, 일치하면 1 이다.

```python
building_area_match_bi = []

for idx, member in df.iterrows():
    if member["Total_building_auction_area"] == member["Total_building_area"] :
        building_area_match_bi.append(1)
    else :
        building_area_match_bi.append(0)

df["building_area_match_bi"] = building_area_match_bi
df["building_area_match_bi"]
    
```

### 4) 강제, 임의경매 구분

임의경매면 0, 강제경매면 1이다.

```python
auction_class_bi = []

for idx, member in df.iterrows():
    if member["Auction_class"] == "임의" :
        auction_class_bi.append(0)
    else :
        auction_class_bi.append(1)

df["auction_class_bi"] = auction_class_bi
df["auction_class_bi"]
```

### 5) 아파트, 주상복합 구분

아파트면 0, 주상복합이면 1이다.

```python
Apartment_usage_bi = []

for idx, member in df.iterrows():
    if member["Apartment_usage"] == "아파트" :
        Apartment_usage_bi.append(0)
    else :
        Apartment_usage_bi.append(1)

df["Apartment_usage_bi"] = Apartment_usage_bi
df["Apartment_usage_bi"]
```

# 과제 2-(3)

회귀계수 계산 함수

```python
from numpy.linalg import inv 

def estimate_beta(x, y):
    """구현해야 하는 부분"""
    beta_hat = np.linalg.inv(x.transpose().dot(x)).dot(x.transpose()).dot(y)
    return beta_hat
```

```python
df["1"] = 1

X_3 = df[["1","Minimum_sales_price"]] # 여러 칼럼을 집어넣으니 역행렬을 구할 수 없다고 떠서 단순회귀로 대체
x_3 = X_3.values
x_3
```

```python
y_3 = df["Hammer_price"].values
y_3
```

```python
beta = estimate_beta(x_3, y_3)
beta
```

MSE 계산 함수

```python
def estimate_mse(beta, x, y):
    square = 0
    for i in range(x.shape[0]):
        y_hat = beta[0]+beta[1]*x[i][1]
        square += np.power((y - y_hat), 2)
    
    mse = square.sum()
    mse = mse/x.shape[0]
    return mse
```

```python
estimate_mse(beta, x_3, y_3)
```
